```python
def cmp(a,b):
    if a['age'] > b['age']:
        return 1
    if a['age'] == b['age']:
        return 0
    if a['age'] < b['age']:
        return -1 
```

```python
def sort(lst, reverse=False):
    dst = []
    for n in lst:
        for i, e in enumerate(dst):
            if not reverse:                 ## 非逆序 
                if cmp(n, e) < 0:
                    dst.insert(i, n)
                    break
            else:
                if cmp(n, e) > 0:
                    dst.insert(i, n)
                    break
        else:
            dst.append(n)
    return dst
```

```python
def sort(lst, cmp=None, reverse=False):     ## 这里定义为None是惯用的用法
    def default_cmp(a, b):  ## 这里定义了一个内部函数，函数可以嵌套定义
        if a > b:           ## 这就就是定义了默认的比较函数
            return 1
        if a == b:
            return 0
        if a < b :
            return -1
    if cmp is None:
        cmp = default_cmp  # 函数是一等对象的一个体现   这里是不能加括号的！！加上就是调用函数了
    dst = []               ## 上一行像变量赋值一样，把一个函数赋值给一个变量
    for n in lst:
        for i, e in enumerate(dst):
            if not reverse:
                if cmp(n, e) < 0:
                    dst.insert(i, n)
                    break
            else:
                if cmp(n, e) > 0:
                    dst.insert(i, n)
                    break
        else:
            dst.append(n)
    return dst
```

```python
def timeit(fn):                             ## 这里是定义装饰器   timeit是装饰器
    def wrap(*args, **kwargs): # 可变参数   定义了一个新的函数wrap
        start = time.time()
        ret = fn(*args, **kwargs)  # 参数解构   这个*args,和**kwargs可以认为是百找搭的参数
        print(time.time() - start)
        return ret
    return wrap
```

```python
def test(a, b):                        ## 定义一个(字典)比较的函数
    if a['age'] > b['age']:
        return 1
    if a['age'] == b['age']:
        return 0
    if a['age'] < b['age']:
        return -1
```

```python
def sort(lst, cmp=None, reverse=False):     ## 这里定义为None是惯用的用法  给cmp一个默认参数为None
    def default_cmp(a, b):  ## 这里定义了一个内部函数，函数可以嵌套定义
        if a > b:           ## 这就就是定义了默认的比较函数
            return 1
        if a == b:
            return 0
        if a < b :
            return -1
    if cmp is None:
        cmp = default_cmp  # 函数是一等对象的一个体现   这里是不能加括号的！！加上就是调用函数了
    dst = []               ## 上一行像变量赋值一样，把一个函数赋值给一个变量，然后还可以调用 
    for n in lst:
        for i, e in enumerate(dst):
            if not reverse:
                if cmp(n, e) < 0:
                    dst.insert(i, n)
                    break
            else:
                if cmp(n, e) > 0:
                    dst.insert(i, n)
                    break
        else:
            dst.append(n)
    return dst
```
