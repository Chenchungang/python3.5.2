```python
class dog:
    def __init__(self,name):
        self.name=name
    def bulk(self):
        print('%s wang wang!'% self.name)
```

### 类变量存在类的内存里

```python
class Role:   ## 定义一个类
    n = 123
    def __init__(self, name, role, weapon, life_value=100, money=15000):

        self.name = name    
        self.role = role   ## 这些都是赋予给实例的  叫实例变量(静态属性)  作用域就是实例本身
        self.weapon = weapon
        self.life_value = life_value
        self.money = money
    def __del__(self):
        print("")

    def shot(self):      ## 角色的功能   专业术语：类的方法 (动态属性)
        print("shooting...")
```

```python
print(Role.n)
```

```python
r1=Role('chen','jingcha','qiang')
```

```python
r1.shot()
```

```python
r1.n
```

```python
r1.n = 'abc'
```

```python
r1.n
```

```python
try:
    print('try...')
    r = 10 / 0
    print('result:', r)
except ZeroDivisionErr as e:
    print('except:', e)
finally:
    print('finally...')
print('END')
```

```python
try:
    print('try...')
    r = 10 / int(0)
    print('result:', r)
except ValueError as e:
    print('ValueError:', e)
except ZeroDivisionError as e:
    print('ZeroDivisionError:', e)
finally:
    print('finally...')
print('END')
```

```python
try:
    print('try...')
    r = 10 / int('2')
    print('result:', r)
except ValueError as e:
    print('ValueError:', e)
except ZeroDivisionError as e:
    print('ZeroDivisionError:', e)
else:
    print('no error!')
finally:
    print('finally...')
print('END')
```

```python
def foo(s):
    return 10 / int(s)

def bar(s):
    return foo(s) * 2

def main():
    bar('0')

main()
```

```python
import logging

def foo(s):
    return 10 / int(s)

def bar(s):
    return foo(s) * 2

def main():
    try:
        bar('0')
    except Exception as e:
        logging.exception(e)

main()
print('END')
```
